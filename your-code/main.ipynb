## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
# your code here
from scipy.stats import uniform
uniform.rvs(2, 8, size=10)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here
# Create uniform distribution function
def generate_uniform(bottom, ceiling, count):
    return uniform.rvs(bottom, ceiling - bottom, size=count)
```

```python
# your code here
uniform_data1 = generate_uniform(10, 15, 100)
uniform_data2 = generate_uniform(10, 60, 1000)

print(f"First dataset (10-15, n=100): mean={uniform_data1.mean():.2f}, std={uniform_data1.std():.2f}")
print(f"Second dataset (10-60, n=1000): mean={uniform_data2.mean():.2f}, std={uniform_data2.std():.2f}")
```

```python
# your code here
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.hist(uniform_data1, bins=10, alpha=0.7, color='blue', edgecolor='black')
ax1.set_title('Uniform Distribution (10-15, n=100)')
ax1.set_xlabel('Value')
ax1.set_ylabel('Count')

ax2.hist(uniform_data2, bins=10, alpha=0.7, color='red', edgecolor='black')
ax2.set_title('Uniform Distribution (10-60, n=1000)')
ax2.set_xlabel('Value')
ax2.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# your answer below
# The two distributions differ in:
# 1. Range: First distribution spans 5 units (10-15), second spans 50 units (10-60)
# 2. Sample size: 100 vs 1000 observations
# 3. The wider range in the second distribution results in lower frequency per bin
# 4. Both maintain uniform probability density within their respective ranges
```

1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
# your code here
def generate_normal(mean, std, count):
    return norm.rvs(loc=mean, scale=std, size=count)
```

```python
# your code here
# Generate normal distributions
normal_data1 = generate_normal(10, 1, 1000)  # mean=10, std=1
normal_data2 = generate_normal(10, 50, 1000)  # mean=10, std=50

print(f"Normal distribution 1 (μ=10, σ=1): mean={normal_data1.mean():.2f}, std={normal_data1.std():.2f}")
print(f"Normal distribution 2 (μ=10, σ=50): mean={normal_data2.mean():.2f}, std={normal_data2.std():.2f}")
```

Which one(s) of the variables are nearly normally distributed? How do you know?

```python
# your code here
# Statistical analysis for normality
from scipy import stats

# Test for normality using Shapiro-Wilk test (sample first 5000 due to size limits)
sample_size = 5000

fuel_sample = vehicles['Fuel Barrels/Year'].sample(sample_size)
co2_sample = vehicles['CO2 Emission Grams/Mile'].sample(sample_size) 
mpg_sample = vehicles['Combined MPG'].sample(sample_size)

# Shapiro-Wilk tests
fuel_stat, fuel_p = stats.shapiro(fuel_sample)
co2_stat, co2_p = stats.shapiro(co2_sample)
mpg_stat, mpg_p = stats.shapiro(mpg_sample)

print("Shapiro-Wilk Normality Tests:")
print(f"Fuel Barrels/Year: statistic={fuel_stat:.4f}, p-value={fuel_p:.2e}")
print(f"CO2 Emissions: statistic={co2_stat:.4f}, p-value={co2_p:.2e}")
print(f"Combined MPG: statistic={mpg_stat:.4f}, p-value={mpg_p:.2e}")

print("\nSkewness values:")
print(f"Fuel Barrels/Year: {stats.skew(vehicles['Fuel Barrels/Year']):.3f}")
print(f"CO2 Emissions: {stats.skew(vehicles['CO2 Emission Grams/Mile']):.3f}")
print(f"Combined MPG: {stats.skew(vehicles['Combined MPG']):.3f}")
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
# Create exponential distribution function and generate data
def generate_exponential(mean_val, size):
    """
    Generate exponentially distributed random numbers
    
    Parameters:
    mean_val (float): Mean of the exponential distribution
    size (int): Number of random numbers to generate
    
    Returns:
    numpy array: Array of exponentially distributed random numbers
    """
    return np.random.exponential(scale=mean_val, size=size)

# Generate two datasets
exp_data1 = generate_exponential(10, 10)
exp_data2 = generate_exponential(10, 100)

# Plot the distributions
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.hist(exp_data1, bins=10, alpha=0.7, color='purple', edgecolor='black')
ax1.set_title('Exponential Distribution (n=10)')
ax1.set_xlabel('Value')
ax1.set_ylabel('Frequency')

ax2.hist(exp_data2, bins=20, alpha=0.7, color='orange', edgecolor='black')
ax2.set_title('Exponential Distribution (n=100)')
ax2.set_xlabel('Value')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()

print(f"Small sample (n=10): mean={exp_data1.mean():.2f}, std={exp_data1.std():.2f}")
print(f"Large sample (n=100): mean={exp_data2.mean():.2f}, std={exp_data2.std():.2f}")
```

How are the two distributions different?

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here
#Bank waiting time problem setup
from scipy.stats import expon

# Exponential distribution with mean = 10 minutes
# For exponential distribution, λ = 1/mean, so λ = 1/10 = 0.1
lambda_rate = 1/10
mean_time = 10

print(f"Bank waiting time follows exponential distribution with mean = {mean_time} minutes")
print(f"Rate parameter λ = {lambda_rate}")
```

```python
# your answer here
# Hint: This is same as saying P(x<15)
# Probability of spending less than 15 minutes
# P(X < 15) for exponential distribution
prob_less_than_15 = expon.cdf(15, scale=mean_time)

print(f"Probability that a customer spends less than 15 minutes: {prob_less_than_15:.4f}")
print(f"This is approximately {prob_less_than_15*100:.1f}%")

# Alternative calculation using the exponential CDF formula: 1 - e^(-λx)
prob_alternative = 1 - np.exp(-lambda_rate * 15)
print(f"Verification using formula 1 - e^(-λx): {prob_alternative:.4f}")
```

What is the probability that the customer will spend more than 15 minutes

```python
# your code here
# Probability of spending more than 15 minutes
# P(X > 15) = 1 - P(X < 15)
prob_more_than_15 = 1 - prob_less_than_15

print(f"Probability that a customer spends more than 15 minutes: {prob_more_than_15:.4f}")
print(f"This is approximately {prob_more_than_15*100:.1f}%")

# Alternative using survival function
prob_survival = expon.sf(15, scale=mean_time)
print(f"Verification using survival function: {prob_survival:.4f}")
```
